```python
df = df.drop(columns=['esearch', 'efetch_responses'], errors='ignore')
df = df.explode('ids')
df = df.rename(columns={'ids': 'pmid'})
df['pmid'] = df['pmid'].astype(float)   # or int if df_one is int
df_one['pmid'].dtype
```

```python
df_merged = df_one.merge(df[['pmid', 'drug_name']], on='pmid', how='left')
```

```python
df_merged = df_merged.dropna(
    subset=['pmid', 'title', 'abstract'],
    how='all'
)
```

```python
df_merged.info()
```

```python
df_merged[df_merged['abstract'].isnull()].head()
```

```python
df_merged['pmid'] = df_merged['pmid'].astype(int)
df_merged = df_merged.dropna(subset=['abstract'])
df_merged = df_merged.drop(columns=[
    'source_jsonl_line',
    'source_retstart',
    'source_efetch_batch',
    'format'
])
```

```python
df_merged.info()
```

```python
df_merged = df_merged.drop_duplicates(subset=['pmid', 'drug_name'])
```

```python
import re

df_merged['drug_name'] = (
    df_merged['drug_name']
    .astype(str)
    .str.lower()
    .str.strip()
)
df_merged['drug_name'] = df_merged['drug_name'].str.strip('.,; ')
```

```python
df_merged['title'] = df_merged['title'].str.strip('.,;[](){}"\' ')
```
